# **Question 01**
An official claims that the average wind speed in the city is 9 mph. Is this plausible ?

```r
qqnorm(airquality$Wind)
```

```r
t.test(airquality$Wind,mu=9,alternative='two.sided',conf.level=0.95)
```

# **Question 02**
A certain solar array will be cost effective only if the mean solar radiation is over 175 Langleys.  Would it ba sound investment based on the available data ?

```r
qqnorm(airquality$Solar.R)
```

```r
t.test(airquality$Solar.R, mu=175,alternative ='greater',conf.level = 0.95)
```
